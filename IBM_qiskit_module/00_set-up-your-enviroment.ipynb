# Get started with Qiskit in the classroom


For this Qiskit in Classrooms module, students must have a working Python environment with the following packages installed:

* `qiskit` v2.1.0 or newer
* `qiskit-ibm-runtime` v0.40.1 or newer
* `qiskit-aer` v0.17.0 or newer
* `qiskit.visualization`
* `numpy`
* `pylatexenc`

To set up and install the packages above, see the [Install Qiskit](https://quantum.cloud.ibm.com/docs/en/guides/install-qiskit) guide.
In order to run jobs on real quantum computers, students will need to set up an account with IBM Quantum® by following the steps in the [Set up your IBM Cloud® account](https://quantum.cloud.ibm.com/docs/en/guides/cloud-setup) guide.

This module was tested and used 2 seconds of QPU time on a Heron v2 processor. This is an estimate only. Your actual usage may vary.

uv is an excellent python manager. Please view the help page with `uv help`

## Use UV now in the commandline
Windows users: make sure you are in your WSL prompt 

Python 3.11 is a decent version supported by qiskit, install it, and install its package manager pip:

```bash
uv python install 3.11
uv pip install pip # Installs python package manager
```

Given you are in the desired location for it (check with `pwd`), make a virtual enviroment for python, and use it:

```bash
uv venv -p 3.11  # Populates $(pwd)/.venv with a python enviroment
source .venv/bin/activate # Runs the 'venv activator' script 
```

Now we are done with uv, and can act like we have a normal python installation. Get the dependencies:

```bash
pip install 'qiskit>=2.1.0' 'qiskit-ibm-runtime>=0.40.1' \
 'qiskit-aer>=0.17.0' 'numpy' 'pylatexenc' 'jupyter' 'matplotlib'
```
